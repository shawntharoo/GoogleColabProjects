* Pandas fillna(): Fill NA/NaN values using the specified method. https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

# Featuretools - EntitySets and Entities
This section will you how to use Featuretools software to generate data features


**STEP 1: tell featuretools what the entityset is**

```python
# the cleaned dataframe
df
```

```python
# es is a varriable to store the result
es = ft.EntitySet(id='diamonds')
```

```python
es
```

**STEP 2: create entity from dataframe**

```python
es  = es.entity_from_dataframe(entity_id='diamonds_data',
                               dataframe=df,
                               make_index = True, 
                               index='index')
```

```python
es
```

**STEP 3: Ask featuretools to use Deep Feature Synthesis (DFS)**

* ft.dfs() creates a features matrix as well as the feature definitions
* ft.dfs() will return two types of data


```python
variable1, variable2 = ft.dfs()
```

```python
# fm stands for features matrix
# fdefs stands for feature definitions
fm, fdefs = ft.dfs(entityset=es, target_entity='diamonds_data')
```

```python
fm
```

```python
fdefs
```

**STEP 4: Normalise**

* this creates relationship

```python
es.normalize_entity(base_entity_id='diamonds_data', new_entity_id='price', index = 'price')
```

```python
es
```

**STEP 5: Generate new features**

* tell featuretools to use a transformation primitive i.e., add and multiply

```python
fm, fdefs = ft.dfs(entityset=es, 
                   target_entity='diamonds_data', 
                   trans_primitives = ['add_numeric', 'multiply_numeric'])
```

```python
fm
```

```python
fdefs
```

```python
plt.figure(figsize=(50,50))
corr = fm.corr()
sns.heatmap(data=corr, square=True , annot=True, cbar=True,linewidth=2)
```

**Remove Highly NULL features**

*  https://featuretools.alteryx.com/en/stable/guides/feature_selection.html


```python
# remove features that are manin zeros
ft.selection.remove_highly_null_features(fm)
```

**Remove Single-valued features**

* remove features that have no variance  (they are uninteresting)

```python
fm_v2, fdefs_v2 = ft.selection.remove_single_value_features(fm, features=fdefs)
fm_v2
```

**Remove Highly-Correlated features**

* The last feature selection function allows us to remove features that is most likely to be redundant to the model we’re attempting to build by considering the correlation between pairs of calculated features.

* When two features are determined to be highly correlated, we remove the more complex of the two. 

```python
fm_v3, fdefs_v3 = ft.selection.remove_highly_correlated_features(fm_v2, features=fdefs_v2)
fm_v3
```

```python
plt.figure(figsize=(50,50))
corr = fm_v3.corr()
sns.heatmap(data=corr, square=True , annot=True, cbar=True,linewidth=2)
```

# Predictions: Price of Diamonds

* in machine learning, the features are all the columns that influences the outcome

```python
# features is all columns apart from what we want to predict - the price
# so we want all the columns apart from the price column
features = fm_v3.loc[:, fm_v3.columns != 'price']
```

```python
features
```

```python
# the label or target or dependent variable  - what  we want to predict
labels = fm_v3.price
```

```python
labels
```

```python
labels.shape
```

```python
# variable X are the features
# variable y are the labels
X = features
y = labels

print(features.head())
print(X.shape)

print(labels.head())
print(y.shape)
```

**ML training and testing dataset**

* use scikit-learn train_test_split to separate dataframe into training and testing


```python
# split data into training and testing sets
# setting random_state to any number ensures
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)

# check how many data points are in the training - X must match y
print('X train:', X_train.shape)
print('y train:', y_train.shape)
```

**Rescale Data**

* ML works best on data between [0,1] or [-1,1]
* so we transform data (standardise)


```python
# call library
scaler = StandardScaler()

# fit scaler to data
X_train_scaler = scaler.fit(X_train)
X_test_scaler = scaler.fit(X_test)

# transform data
X_train_scaled = X_train_scaler.transform(X_train)
X_test_scaled = X_test_scaler.transform(X_test)
```

```python
print(X_train_scaled)
```

# Linear Regression

*  https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html


```python
# create a linear regression model
linear = LinearRegression()
```

```python
# show the linear model that was created
linear
```

```python
# train the linear model on training data
linear.fit(X_train_scaled, y_train)
```

```python
# use the trained linear model to make predictions
linear_prediction = linear.predict(X_test_scaled)
```

```python
linear_prediction
```

```python
# initialise the model
lreg_model = LinearRegression()


# train the model on training data
lreg_model.fit(X_train_scaled,y_train)
```

```python
pred = lreg_model.predict(X_test_scaled)
```

```python
# make predictions on test data
prediction = lreg_model.predict(X_test_scaled)

# calculate the root mean square error - a simple standard metric for accuracy
rmse_lreg = np.sqrt(mean_squared_error(y_test, prediction))
print('RMSE value is = {}'.format(rmse_lreg))
```

**Accuracy Metric: R-square value**

*  https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html


```python
# calculate r-squared value - how accurate
r2_lreg = r2_score(y_test, prediction)
print('R-squared value is {}'.format(r2_lreg))
```

```python
lreg_score_train = lreg_model.score(X_train_scaled, y_train)
print(lreg_score_train)
```

```python
lreg_score_test = lreg_model.score(X_test_scaled, y_test)
print(lreg_score_test)
```

# Random Forest

* A random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

* https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

```python
# create the model
rf_model = RandomForestRegressor()

# train the model
rf_model.fit(X_train_scaled,y_train)
```

```python
# make predictions
predictions_rf = rf_model.predict(X_test_scaled)

# calculate rmse
rmse_rf = np.sqrt(mean_squared_error(y_test, predictions_rf))
print('RMSE value is = {}'.format(rmse_rf))
```

```python
# calculate r-square metric
r2_rf = r2_score(y_test, predictions_rf)
print('R-squared value is {}'.format(r2_rf))
```

```python
rf_score_train = rf_model.score(X_train_scaled, y_train)
print(rf_score_train)
```

```python
rf_score_test = rf_model.score(X_test_scaled, y_test)
print(rf_score_test)
```
