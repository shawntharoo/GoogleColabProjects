As seen in the het map all the means and the worst values are highly corelated. also paramter mean and radius mean also highly corelated. therefore we can remove all the worst values and parameter mean and worst columns from the dataset.

Since above values are strongly corelated replace with a single value in the "Calculated" column

```python
df['radius_calculated'] = df['radius_worst'] * df['radius_mean'] 
```

```python
df['texture_calculated'] = df['texture_worst'] * df['texture_mean']
```

```python
df['perimeter_calculated'] = df['perimeter_worst'] * df['perimeter_mean']
```

```python
df['area_calculated'] = df['area_worst'] * df['area_mean']
```

```python
df['smoothness_calculated'] = df['smoothness_worst'] * df['smoothness_mean']
```

```python
df['concavity_calculated'] = df['concavity_worst'] * df['concavity_mean']
```

```python
df['concave points_calculated'] = df['concave points_worst'] * df['concave points_mean']
```

```python
df['symmetry_calculated'] = df['symmetry_worst'] * df['symmetry_mean']
```

```python
df['compactness_calculated'] = df['compactness_worst'] * df['compactness_mean']
```

```python
df['fractal_dimension_calculated'] = df['fractal_dimension_worst'] * df['fractal_dimension_mean']
```

```python
df['ar_peri_rad_calculated'] = df['radius_calculated'] * df['perimeter_calculated'] * df['area_calculated']
```

```python
df['ar_peri_rad_calculated_SE'] = df['radius_se'] * df['perimeter_se'] * df['area_se']
```

```python
dropcolumns = ['radius_worst',
              'texture_worst',
              'perimeter_worst',
              'area_worst',
              'smoothness_worst',
              'compactness_worst',
              'concavity_worst',
              'concave points_worst',
              'symmetry_worst',
              'fractal_dimension_worst', 
              'radius_mean',
              'texture_mean',
              'area_mean',
              'smoothness_mean',
              'compactness_mean',
              'concavity_mean',
              'concave points_mean',
              'symmetry_mean',
              'fractal_dimension_mean',
              'perimeter_mean',
              'radius_calculated',
              'perimeter_calculated',
              'area_calculated',
              'radius_se',
              'perimeter_se',
              'area_se']
```

```python
df = df.drop(dropcolumns, axis=1)
```
